## **CNN with CIFAR-10 Data**

```python
import pandas as pd
import seaborn as sns
import tensorflow as tf                       
import numpy as np                           
import matplotlib.pyplot as plt               
%matplotlib inline
```

```python
from tensorflow.keras.datasets import cifar10  # keras ta var cifar 10 dedik, tuple unpacking yaptık train test split yapmadık.

(X_train, y_train), (X_test, y_test) = cifar10.load_data()
```

## Data Preprocessing

```python
X_train[5].min()
```

```python
X_train[5].max()
```

```python
X_train = X_train.astype('float32')
X_test = X_test.astype('float32')
X_train /= 255
X_test /= 255  # piksel değerlerini 0 -1 arasına topladık NORMALİZE işlemi
```

```python
X_train[5].min()
```

```python
X_train[5].max()  # 99 a yakın bir rakam oldu
```

```python
plt.imshow(X_train[5])
```

```python
X_train.shape, X_test.shape
```

```python
y_train  # sadece int. rakamlar var bunları categoric yapmalı
```

```python
from tensorflow.keras.utils import to_categorical
# cat. import ettik. verileri 10 clas olarak categoric yaptık
# int 1 artık cat 1 olacak. İNT--->CATEGORİC yaptık
```

```python
Y_train = to_categorical(y_train, 10) 
Y_test = to_categorical(y_test, 10)
Y_test.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Conv2D, MaxPool2D, Flatten, Dropout # over fite gitmesin dedik dropout
```

```python
model = Sequential()  # model oluştu

model.add(Conv2D(filters=32, kernel_size=(4,4),input_shape=(32, 32, 3), activation='relu',))
# filters=32, kernel_size=(4,4),input_shape=(32, 32, 3), activation='relu'
model.add(MaxPool2D(pool_size=(2, 2)))

model.add(Conv2D(filters=32, kernel_size=(4,4),input_shape=(32, 32, 3), activation='relu',))
# bir kez daha 
model.add(MaxPool2D(pool_size=(2, 2)))

model.add(Flatten()) # vekör haline getirdik

model.add(Dense(256, activation='relu')) # dense rank 256 128 de olur 512 de

model.add(Dense(10, activation='softmax'))

model.compile(loss='categorical_crossentropy',
              optimizer='adam',      # diğer leride olabilir.
              metrics=['accuracy'])  # model fit olurken takip edilir.
# fully connected olmayınca değerler düştü daraldı.
```

```python
model = Sequential()  # model oluştu

model.add(Conv2D(filters=64, kernel_size=(8,8),input_shape=(32, 32, 3), activation='relu',))
# filters=32, kernel_size=(4,4),input_shape=(32, 32, 3), activation='relu'
model.add(MaxPool2D(pool_size=(2, 2)))

model.add(Conv2D(filters=64, kernel_size=(8,8),input_shape=(32, 32, 3), activation='relu',))
# bir kez daha 
model.add(MaxPool2D(pool_size=(2, 2)))

model.add(Flatten()) # vekör haline getirdik

model.add(Dense(512, activation='relu')) # dense rank 256 128 de olur 512 de

model.add(Dense(10, activation='softmax'))

model.compile(loss='categorical_crossentropy',
              optimizer='adam',      # diğer leride olabilir.
              metrics=['accuracy'])  # model fit olurken takip edilir.
# fully connected olmayınca değerler düştü daraldı.
```

```python
model.summary()
```

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
early_stop = EarlyStopping(monitor='val_loss',patience=2)  # model 2 epochs tta dursun, patience=2 sabır demek 2 defa da aynıysa dur.
```

```python
model.fit(X_train, Y_train, batch_size=32, epochs=20, validation_data=(X_test,Y_test),callbacks=[early_stop])
                                                     #  validation_data da testleri kullandık.
```

```python
model.metrics_names
```

```python
model.history.history
```

```python
summary = pd.DataFrame(model.history.history)
summary.head()
```

```python
plt.figure(figsize=(10,6))
plt.plot(summary.loss, label="loss")
plt.plot(summary.val_loss, label="val_loss")
plt.legend(loc="upper right")
plt.ylabel("Loss")
plt.xlabel("Epoch")
plt.show()
# 3 te kesilebilir. over fittinge gitmesin
```

```python
plt.figure(figsize=(10,6))
plt.plot(summary.accuracy, label="accuracy")
plt.plot(summary.val_accuracy, label="val_accuracy")
plt.legend(loc="upper left")
plt.ylabel("Accuracy")
plt.xlabel("Epoch")
plt.show()
# arada açıklık varsa overfit gider 3 te sonlansa iyi
```

```python
print(model.evaluate(X_test,Y_test,verbose=1))
```

```python
from sklearn.metrics import classification_report,confusion_matrix, plot_confusion_matrix
```

```python
preds = model.predict(X_test)
```

```python
predictions= np.argmax(preds, axis=1) # x ekseni boyunca en yuksek değerli datanın indexini veriyor o da class oldu
```

```python
predictions
```

```python
print(classification_report(y_test, predictions))
```

```python
print(confusion_matrix(y_test, predictions))
```

```python
score = model.evaluate(X_test, Y_test, verbose=0)
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

```python
model.save("cnn-1.h5")  # kaydetmek için bu acary çok değil ama, 
# early stop ekelrsek bazen accuary veya loos düşmeden de model kesebilir.
```

```python
model= Sequential()
model.add(Conv2D(filters=32,kernel_size=(3,3),activation='relu',input_shape=(32,32,3)))
model.add(MaxPool2D((2,2)))
        
model.add(Conv2D(filters=64,kernel_size=(3,3),activation='relu')) # 2. de kernel sayısını 64 yaptık
model.add(MaxPool2D((2,2)))
model.add(Dropout(0.3))     # random 30  yuzde kapat
    
model.add(Conv2D(filters=128,kernel_size=(3,3),activation='relu'))
model.add(MaxPool2D((2,2)))
model.add(Dropout(0.3))     # random 30  yuzde kapat
        
model.add(Flatten())
model.add(Dense(512, activation='relu'))  # parametre sayısna göre artması iyi olur
model.add(Dense(10, activation='softmax'))

model.compile(loss='categorical_crossentropy',  # category crossentropy
              optimizer='adam',
              metrics=['accuracy'])
#Total params: 361,034 oldu.
```

```python
model.summary()
```

```python
model.fit(X_train, Y_train, batch_size=32, epochs=30, validation_data=(X_test,Y_test))
# Early stop edersem bazen modelin önünü keserim.
```

```python
score = model.evaluate(X_test, Y_test, verbose=0)
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

```python
preds = model.predict(X_test)
predictions= np.argmax(preds, axis=1)
```

```python
print(classification_report(y_test, predictions))
```

```python
print(confusion_matrix(y_test, predictions))
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(confusion_matrix(y_test, predictions), annot=True)
# buradan da bakılabilir.
```
